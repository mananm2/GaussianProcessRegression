```python
import numpy as np
import pandas as pd
import scipy.io
import matplotlib.pyplot as plt
```

```python
for i in range(1,21):    
    vec = scipy.io.loadmat('EngineDataset/part_'+str(i)+'_vec.mat')
    arr = vec['xyz']
    df = pd.DataFrame(arr, columns = ['X', 'Y','Z'])

    #mask to select every nth point (here n is 5)
    n = 5
    mask = [True if i%n==0 else False for i in range(len(df))]

    #Alter the dataframe
    df = df[mask]

    #Save as csv
    df.to_csv('EngineDatasetShrinkedCSV/part_'+str(i)+'_'+str(n)+'.csv', index = False)
```
